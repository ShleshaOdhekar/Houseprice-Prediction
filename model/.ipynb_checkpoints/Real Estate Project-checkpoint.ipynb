```python
def convert_sqft(x):
    tokens= x.split('-')
    if len(tokens)==2:
        mean= (float(tokens[0])+float(tokens[1]))/2
        return mean
    try:
        return float(x)
    except:
        return None
    
            
```

```python
df6.shape
```

```python
def remove_ppsf_outliers(df):
    df_out= pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m= np.mean(subdf['price_per_sqft'])
        st= np.std(subdf['price_per_sqft'])
        red_df= subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<(m+st))]
        df_out=pd.concat([df_out, red_df], ignore_index=True)
    return df_out
```

### Now, let's remove those 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment



```python
def remove_bhk_outliers(df):
    new_df=pd.DataFrame()
    exclude_indices=([])
    for name, loc_df in df.groupby('location'):
        bhk_stats={}
        for bno, b_df in loc_df.groupby('bhk'):
            bhk_stats[bno] = {
                'mean': np.mean(b_df.price_per_sqft),
                'std': np.std(b_df.price_per_sqft),
                'count': b_df.shape[0]
            }
        for bno, b_df in loc_df.groupby('bhk'):
            #filters 2 bhk
            stats= bhk_stats.get(bno-1)
            if stats and stats['count']> 5:
                #rint(b_df)
                exclude_indices= np.append(exclude_indices,b_df[(b_df.price_per_sqft)<(stats['mean'])].index.values)
        #rint(name)
        #rint(exclude_indices)
    return df.drop(exclude_indices, axis='index')
                
                                
            
       
```

```python
len(Y)
```

```python
#using linearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=10 )
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train, Y_train)
lr.score(X_test, Y_test)
```

```python
#using kfold to come up with best optimal model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv= ShuffleSplit(n_splits= 5, test_size=0.2, random_state=0 )
cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
#using grid_search for hyperparameter tuning to get the best parameters for a particular model
#then use kfold to decide which model among them is best accurate
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
def using_gridsearch(X, Y):
    algos= {
        'linear Regression':
        {
          'model': LinearRegression(),
           'params': {
               'normalize': [True, False]
            },  
        },
        'Lasso':
        {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            },
        },
        'Decision Tree':
        {
            'model' : DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    #print(algos.items())
    scores=[]
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, model in algos.items():
        #print(model['model'])
        
        gs =  GridSearchCV(model['model'], model['params'], cv=cv, return_train_score=False)
        gs.fit(X,Y)
        
        scores.append({
            'model': algo_name,
            'score': gs.best_score_,
            'params': gs.best_params_
        })
    return pd.DataFrame(scores)
        
using_gridsearch(X, Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=10 )
lr_clf= LinearRegression(normalize=True)
lr_clf.fit(X_train, Y_train)
lr_clf.score(X_test, Y_test)
```
